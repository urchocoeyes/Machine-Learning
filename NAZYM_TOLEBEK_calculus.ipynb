## Task 3.2 (0.5 points)

Let $\boldsymbol X \in \mathbb R^{m\times n}$ and $f(\boldsymbol X) = \vert \boldsymbol X \vert^k$, $k\in\mathbb N$. Find $\nabla f(\boldsymbol X)$.

### MY SOLUTION

$$ f(\mathbf{X}) = |\mathbf{X}|^k $$

Where $ |\mathbf{X}| $ denotes the absolute value taken element-wise, and $ k $ is a positive integer.

To find the gradient of $ f(\mathbf{X}) $, we'll differentiate it with respect to each element of $ \mathbf{X} $ (each $ X_{ij} $).

The gradient with respect to $\boldsymbol X$ can be calculated using the chain rule. We have:

$$ \nabla f(\mathbf{X}) = k \cdot \mathbf{|\boldsymbol X |}^{k-1} \cdot \nabla \mathbf{|X|} $$

Let;s find $\nabla \mathbf{|X|} $, which represents the gradient of $ |\mathbf{X}| $ with respect to $\mathbf{X}$.

The absolute value function applied to each element is a piecewise function, but its derivative is well-defined. The derivative of the absolute value function $ |\boldsymbol X|$  with respect to its element $x$ is:

$$ \frac{d}{dx} |x| = \begin{cases} 1 & \text{if } x > 0 \\ -1 & \text{if } x < 0 \\ \text{undefined} & \text{if } x = 0 \end{cases} $$

So, the derivative is either $ 1 $ or $-1 $ depending on the sign of $ x $, and it's not defined at $ x = 0 $.

$$ \partial |X_{ij}| = \begin{cases} 1 & \text{if } X_{ij} > 0 \\ \text {undefined} & \text{if } X_{ij} = 0 \\ -1 & \text{if } X_{ij} < 0 \end{cases} $$

Putting it all together, we have:

$$ \nabla f(\mathbf{X})_{ij} =
 k \cdot |\mathbf{X}_{ij}|^{k-1} \cdot \begin{cases} 1 & \text{if } \mathbf{X}_{ij} > 0 \\ -1 & \text{if } \mathbf{X}_{ij} < 0 \end{cases} $$

This expression represents the gradient of $ f(\mathbf{X}) $ with respect to each element of $ \mathbf{X} $.

    $$

### MY SOLUTION:
The Frobenius norm of a matrix is a way to measure the "size" or "magnitude" of a matrix. For a matrix $\boldsymbol A\boldsymbol X -\boldsymbol B$ with elements $(\boldsymbol A\boldsymbol X -\boldsymbol B)_{ij}$, the Frobenius norm, denoted as $\Vert\boldsymbol A\boldsymbol X -\boldsymbol B\Vert_{F}$, is defined as the square root of the sum of the squares of all the elements:
$$
\Vert\boldsymbol A \boldsymbol X - \boldsymbol B\Vert_{F}= \sqrt{\sum_{i=1}^k\sum_{j = 1}^n(\boldsymbol A \boldsymbol X - \boldsymbol B)_{ij}^2}
$$
Where $k$ - number of rows, after finding difference,
<br> 
$n$ - number of columns, after finding difference.
<br>
We are aimed to find such $\boldsymbol X$ that minimizes the Frobenius norm of the difference $\boldsymbol A\boldsymbol X -\boldsymbol B$:
$$
\min\limits_{\boldsymbol X}\Vert \boldsymbol A \boldsymbol X - \boldsymbol B \Vert_F^2 = \min\limits_{\boldsymbol X}\sum_{i=1}^{k}\sum_{j=1}^{n}(\boldsymbol A \boldsymbol X - \boldsymbol B )_{ij}^{2}
$$  
To minimize the $\boldsymbol X$, we need to take the derivative of $\Vert \boldsymbol A \boldsymbol X - \boldsymbol B \Vert_{F}^2$  with respect to $\boldsymbol X$ and set it to 0:
<br>
$$
\\
\frac{\partial \Vert \boldsymbol A \boldsymbol X - \boldsymbol B \Vert_{F}^2}{\partial \boldsymbol X} 
= 
\\
\frac {\partial}{\partial \boldsymbol X} \sum_{i = 1}^k\sum_{j=1}^n(\boldsymbol A \boldsymbol X - \boldsymbol B)_{ij}^2 = 
\\
\sum_{i=1}^k\sum_{j=1}^n\frac{\partial}{\partial \boldsymbol X}(\boldsymbol A\boldsymbol X - \boldsymbol B)_{ij}^2 = 
\\
\sum_{i=1}^k
\sum_{j=1}^n 2(\boldsymbol A\boldsymbol X - \boldsymbol B)_{ij}\frac{\partial}{\partial \boldsymbol X}(\boldsymbol A\boldsymbol X - \boldsymbol B)_{ij} =
\\
2 \sum_{i=1}^k\sum_{j=1}^n(\boldsymbol A\boldsymbol X-\boldsymbol B)_{ij}(\boldsymbol A)_{ij}
$$
Since, $\boldsymbol B$ and $\boldsymbol A$ are treated as constants for partial derivative with respect to $\boldsymbol X$.
**The derivative of $\Vert \boldsymbol A \boldsymbol X - \boldsymbol B \Vert_{F}^2$  with respect to $\boldsymbol X:$**
$$
\frac{\partial \Vert \boldsymbol A \boldsymbol X - \boldsymbol B \Vert_{F}^2}{\partial \boldsymbol X} = 2 \sum_{i=1}^k\sum_{j=1}^n(\boldsymbol A\boldsymbol X-\boldsymbol B)_{ij}(\boldsymbol A)_{ij}
$$
Set this derivative to 0 in order to find $\boldsymbol X$:
$$
2 \sum_{i=1}^k\sum_{j=1}^n(\boldsymbol A\boldsymbol X-\boldsymbol B)_{ij}(\boldsymbol A)_{ij} = 0
